read, readline, readlines

```python
#read
file = open('sample.txt', 'r')

print(file.read())
file.close()
```

```python
#readline
file = open('sample.txt', 'r')
print(file.readline())
file.close()
```

```python
#readlines
file = open('sample.txt', 'r')
print(file.readlines())
file.close()
```

```python
file = open('sample.txt', 'r')
txt = file.readlines()
print(txt[1])
file.close()
```

```python
with open('sample.txt', 'r') as file:
    print(file.read())
```

```python
#Seek and tell function into the file handling
with open('sample.txt', 'r') as f:
    # Reading the file line of the file
    # print(f.readline())

    #finging the pointer
    print(f.tell())
    # Seeking the pointer to 0
    print(f.seek(10))
    print(f.readline())
```

```python
# writeline
#Take big files

big_data = ['hello_world\n' for i in range(2000)]
with open('new.txt', 'w') as f:
    f.writelines(big_data)
```

```python
# Reading the big files into the small chunks
with open('new.txt', 'r') as f:
    chunk_size = 100
    while True:
        chunk = f.read(chunk_size)
        if not chunk:
            break
        print("chunk ->")
        print(chunk)
```

```python
# r+ mode in file handling
# this is used for the reading and writing the file but it works for existing file only
#w
with open('new.txt', 'r+') as f:
    f.write('anything')
    print(f.tell())
    f.seek(0)
    print(f.read())
```
