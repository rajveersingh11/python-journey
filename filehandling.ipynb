```python
#Opening a file
file = open('demo.txt', 'r')
print(file.read())
file.close()
```

```python
# Writing something
file = open('demo.txt', 'w')
file.write('this is new line')
file.close()
```

```python
#if any txt is aleady present into the file then we use append new line using 'a'.
file = open('demo.txt','a')
file.write("\n This is append txt")
file.write("\n This is new line")
file.close()
```

```python
#if any file doesn't exist
file = open('sample.txt','w')
file.write('this is new file')
file.close()
```

```python
sentences = ['This is the first line from the list', '\nThis is the second line from the list']
file = open('sample.txt', 'w')
for sent in sentences:
    file.write(sent)
file.close()    
```

read, readline, readlines

```python
#read
file = open('sample.txt', 'r')

print(file.read())
file.close()
```

```python
#readline
file = open('sample.txt', 'r')
print(file.readline())
file.close()
```

```python
#readlines
file = open('sample.txt', 'r')
print(file.readlines())
file.close()
```

```python
file = open('sample.txt', 'r')
txt = file.readlines()
print(txt[1])
file.close()
```

```python
with open('sample.txt', 'r') as file:
    print(file.read())
```
